```python
import pandas as pd


def clean_data(df):
    # Select columns: 'city', 'country' and 2 other columns
    df = df.loc[:, ["city", "country", "iso3", "admin_name"]]
    # Drop duplicate rows across all columns
    df = df.drop_duplicates()
    # Sort by columns: 'country' (ascending), 'admin_name' (ascending), 'city' (ascending)
    df = df.sort_values(["country", "admin_name", "city"], na_position="first")
    df["combined_city"] = df[["city", "admin_name"]].apply(
        lambda row: (row["city"] + ", " + row["admin_name"])
        if pd.notna(row["admin_name"])
        else row["city"],
        axis=1,
    )
    # Select columns: 'country', 'iso3', 'combined_city'
    df = df.loc[:, ["country", "iso3", "combined_city"]]
    # Drop duplicate rows across all columns
    df = df.drop_duplicates()
    return df


# Loaded variable 'df' from URI: d:\.dev\.repo\social-media-platform-v2\database\worldcities.csv
df = pd.read_csv(r"d:\.dev\.repo\social-media-platform-v2\database\worldcities.csv")

df_clean = clean_data(df.copy())
df_clean.head()
```
